```python
import pandas as pd
import math as math
import numpy as np
import matplotlib.pyplot as plt

# Importar módulos personalizados
from Funciones.Constantes import *
from Funciones.Metodos import *
from Funciones.Errores import *
from Funciones.Graficas import *

# Funciones requeridas para este apartado
def y1(t):
    """Solución analítica: y(t) = y0 * e^(-λt)"""
    return y0 * np.exp(-LAMBDA * t)

def dy1(t, y):
    """Derivada de la EDO: dy/dt = -λy"""
    return -LAMBDA * y
```

```python


# Cargar la solución exacta
valores_t = np.linspace(T_INICIAL,T_FINAL,100)
valores_solucion_exacta_grafica = [y1(t) for t in valores_t]

# Graficar solucion exacta
graficar_solucion_exacta(valores_t, valores_solucion_exacta_grafica)

# Valores para pruebas con distintos h
t_inicial = T_INICIAL
t_final = T_FINAL
valores_h = VALORES_H
valores_t = np.linspace(t_inicial-0.5, t_final+0.5, 100)
valores_solucion_exacta = [y1(t) for t in valores_t]

print("== Pruebas con distintos valores de h ==")

for h in valores_h:
    
    print(">>>>>>>>>>>>>>>>")
    print(">>>>> Con h = ", h)
    print(">>>>>>>>>>>>>>>>")

    # Apartado a) RK2
    print("== Runge-Kutta 2 (Trapecio) ==")
    tabla_valores_1_rk2 = runge_kutta_2_trapecio(t_inicial, t_final, h, y1(t_inicial), dy1)
    calcular_y_reales(tabla_valores_1_rk2, y1)
    calcular_error_global(tabla_valores_1_rk2)
    display(tabla_valores_1_rk2)

    # Apartado a) RK4
    print("== Runge-Kutta 4 ==")
    tabla_valores_1_rk4 = runge_kutta_4(t_inicial, t_final, h, y1(t_inicial), dy1)
    calcular_y_reales(tabla_valores_1_rk4, y1)
    calcular_error_global(tabla_valores_1_rk4)
    display(tabla_valores_1_rk4)

    # Apartado a) Adam-Bashforth 2
    print("== Adam-Bashforth 2 ==")
    tabla_valores_1_ab2 = adam_bashforth_2_con_euler_RK1_explicito(t_inicial, t_final, h, y1(t_inicial), dy1)
    calcular_y_reales(tabla_valores_1_ab2, y1)
    calcular_error_global(tabla_valores_1_ab2)
    display(tabla_valores_1_ab2)

    # Apartado a) Heun
    print("== Heun ==")
    tabla_valores_1_heun = heun(t_inicial, t_final, h, y1(t_inicial), dy1, 0.01, 5)
    calcular_y_reales(tabla_valores_1_heun, y1)
    calcular_error_global(tabla_valores_1_heun)
    display(tabla_valores_1_heun)

    # Métodos de scipy
    print("== RK45 (scipy) ==")
    tabla_valores_1_rk45 = resolver_con_scipy('RK45', dy1, t_inicial, t_final, y1(t_inicial))
    calcular_y_reales(tabla_valores_1_rk45, y1)
    calcular_error_global(tabla_valores_1_rk45)
    display(tabla_valores_1_rk45)

    print("== RK23 (scipy) ==")
    tabla_valores_1_rk23 = resolver_con_scipy('RK23', dy1, t_inicial, t_final, y1(t_inicial))
    calcular_y_reales(tabla_valores_1_rk23, y1)
    calcular_error_global(tabla_valores_1_rk23)
    display(tabla_valores_1_rk23)

    print("== DOP853 (scipy) ==")
    tabla_valores_1_dop853 = resolver_con_scipy('DOP853', dy1, t_inicial, t_final, y1(t_inicial))
    calcular_y_reales(tabla_valores_1_dop853, y1)
    calcular_error_global(tabla_valores_1_dop853)
    display(tabla_valores_1_dop853)

    print("== Radau (scipy) ==")
    tabla_valores_1_radau = resolver_con_scipy('Radau', dy1, t_inicial, t_final, y1(t_inicial))
    calcular_y_reales(tabla_valores_1_radau, y1)
    calcular_error_global(tabla_valores_1_radau)
    display(tabla_valores_1_radau)

    print("== BDF (scipy) ==")
    tabla_valores_1_bdf = resolver_con_scipy('BDF', dy1, t_inicial, t_final, y1(t_inicial))
    calcular_y_reales(tabla_valores_1_bdf, y1)
    calcular_error_global(tabla_valores_1_bdf)
    display(tabla_valores_1_bdf)

    print("== LSODA (scipy) ==")
    tabla_valores_1_lsoda = resolver_con_scipy('LSODA', dy1, t_inicial, t_final, y1(t_inicial))
    calcular_y_reales(tabla_valores_1_lsoda, y1)
    calcular_error_global(tabla_valores_1_lsoda)
    display(tabla_valores_1_lsoda)

    # Crear figura completa con todas las gráficas usando función modular
    crear_figura_completa_comparacion(h, valores_t, valores_solucion_exacta,
                                     tabla_valores_1_rk2, tabla_valores_1_rk4, 
                                     tabla_valores_1_ab2, tabla_valores_1_heun,
                                     tabla_valores_1_rk45, tabla_valores_1_rk23,
                                     tabla_valores_1_dop853, tabla_valores_1_radau,
                                     tabla_valores_1_bdf, tabla_valores_1_lsoda)
    plt.show()



```
